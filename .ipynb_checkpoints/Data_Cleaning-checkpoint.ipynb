```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv("train_values.csv", delimiter=",")
train_label = pd.read_csv("train_labels.csv", delimiter=",")
```

```python
test_data = pd.read_csv("test_values.csv", delimiter=",")
```

```python
train_data.head()
```

```python
def explore_feature(feature_name, df):
    val_counts = df[feature_name].value_counts()
    val_counts.plot(kind='bar')
    plt.title(feature_name)
    plt.show()
```

```python
def explore_corr(feature, label_feature, df):
    correlation_df = df[[feature]]
    
    if isinstance(df[feature].dtype, object):
        correlation_df = pd.get_dummies(correlation_df)
        
    correlation_df['label'] = df[label_feature]
    
    #print(correlation_df.corr())
    
    #plotting
    plt.figure()
    plt.title(feature)
    cmap = sns.diverging_palette(h_neg=10, h_pos=240, as_cmap=True)
    sns.heatmap(correlation_df.corr(), center=0, cmap=cmap, linewidths=1, annot=True, fmt=".2f",  xticklabels=False)
```

```python
def full_data():
    damage_grade = train_label['damage_grade']
    full_data = train_data 
    full_data['damage_grade'] = damage_grade
    return full_data
```

```python
categorical_features =[ 'land_surface_condition', 'foundation_type', 'roof_type',                               
                        'ground_floor_type', 'other_floor_type', 'position',                                
                        'plan_configuration', 'legal_ownership_status'
                      ]
```

```python
for feature in categorical_features:
    explore_feature(feature, train_data)
```

```python
data = full_data()
```

```python
for feature in categorical_features:
    explore_corr(feature, 'damage_grade', data)
```

```python
explore_corr('area_percentage', 'damage_grade', data)
```
