```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv("train_values.csv", delimiter=',')
```

```python
train_labels = pd.read_csv("train_labels.csv", delimiter=",")
```

```python
train_damage_grade = train_labels['damage_grade']

information = train_data['geo_level_2_id']
```

```python
full_data = train_data
full_data['damage_grade'] = train_damage_grade
```

```python
full_data.head()
```

```python
df = pd.DataFrame({'information':information, 'damage_grade':train_damage_grade})
```

```python
grouped = df.groupby(['information']).damage_grade.agg([sum, len])
```

```python
grouped.head()
```

```python
grouped['mean'] = (grouped['sum'] // grouped['len'])
```

```python
features = ['age','area_percentage','height_percentage', 'has_secondary_use_institution'  , 'damage_grade']

compare_df = full_data[features]
```

```python
cmap = sns.diverging_palette(h_neg=10, h_pos=240, as_cmap=True)
sns.heatmap(compare_df.corr(), center=0,cmap=cmap, linewidths=1, annot=True, fmt=".2f")
```
